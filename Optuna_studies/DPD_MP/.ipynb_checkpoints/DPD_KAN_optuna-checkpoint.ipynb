```python
def RoF_channel(sigTx, G_list, Rs, SpS, Fs, paramMZM, paramChannel, paramRF, paramPD):
    G1, G2, G3 = G_list
    
    # Sinal RF
    t = np.arange(0, len(sigTx))*1/Fs
    sigTx_RF = np.real( sigTx * np.exp(1j*2*pi*paramRF.fc_e*t) )
    sigTx_RF *= G_1
    
    # Sinal óptico
    Ai     = np.sqrt(dBm2W(paramMZM.Pin_OF))*np.ones(sigTx_RF.size)
    sigTxo = mzm(Ai, sigTx_RF, paramMZM)
    
    numtaps = 4096
    hopt_tx = firwin(numtaps, paramRF.fc_e + 2*Rs, fs = Fs)
    sigTxo = np.sqrt(signal_power(sigTxo))*pnorm(firFilter(hopt_tx, sigTxo))
    
    sigRxo = linearFiberChannel(sigTxo, paramChannel)
    
    # Sinal elétrico (fotocorrente)
    I_Rx = photodiode(sigRxo, paramPD)
    I_Rx -= I_Rx.mean()
    
    # Sinal elétrico pós-FPF
    numtaps = 4096
    f1 = paramRF.fc_e - 2*Rs
    f2 = paramRF.fc_e + 2*Rs
    hbp_RF = firwin(numtaps, (f1, f2), pass_zero = 'bandpass', fs = Fs)
    
    I_RF = firFilter(hbp_RF, I_Rx)
    
    # Sina elétrico pós-PA
    sigRx = hilbert(I_RF)*np.exp(-1j*2*pi*paramRF.fc_e*t)
    sigRx *= G_2
    
    sigRx = powerAmplifier(sigRx)
    
    sigRx_PA = sigRx.copy()
    sigRx_PA *= G_3
    
    numtaps = 4096
    hlp = firwin(numtaps, Rs/1.5, fs = Fs)
    #sigRx = firFilter(hlp, sigRx)
    
    delay = finddelay(sigRx, sigTx)
    sigRx = np.roll(sigRx, -delay)
    
    rot = np.mean(sigTx/sigRx)
    sigRx = rot/np.abs(rot)*sigRx

    # Parâmetros da decimação
    paramDec = parameters()
    paramDec.SpS_in  = SpS
    paramDec.SpS_out = 1
    
    symbRx_OFDM = decimate(sigRx, paramDec).ravel()
    symbRx = demodulateOFDM(symbRx_OFDM, paramOFDM)
    
    return sigRx_PA, symbRx
```

### 1 - Geração do Sinal OFDM

```python
device = "cuda"
includeMemory = True

# Parâmetros de modulação
modOrder = 64                          # Ordem da modulação
modType  = 'qam'                       # Tipo de modulação
Rb = 5e9                               # Taxa de transmissão de bits
SpS = 32                               # Amostras por símbolo

# Parâmetros OFDM
paramOFDM = parameters()
paramOFDM.Nfft = 1024                 # Tam. da FFT
paramOFDM.G    = 32                   # Tam. do prefixo cíclico
paramOFDM.hermitSymmetry = False
paramOFDM.SpS = SpS

if paramOFDM.hermitSymmetry:
    Ns = paramOFDM.Nfft//2 - 1
    paramOFDM.pilotCarriers = np.linspace(0, Ns - 1, 8, dtype = np.int64)
    paramOFDM.nullCarriers  = np.array([], dtype = np.int64)
else:
    Ns = paramOFDM.Nfft
    paramOFDM.pilotCarriers = np.linspace(0, Ns - 1, 8, dtype = np.int64)
    paramOFDM.nullCarriers  = np.array([paramOFDM.Nfft//2], dtype = np.int64)

# Núm. de subportadoras de informação
Np = paramOFDM.pilotCarriers.size
Nz = paramOFDM.nullCarriers.size
Ni = Ns - Np - Nz

Rs = Rb / ( Ni/(paramOFDM.Nfft + paramOFDM.G) * np.log2(modOrder) )    # Taxa de transmissão de símbolos
Fs  = Rs * SpS                                                          # Taxa de amostragem

# Geração dos bits e dos símbolos QAM
numOFDMframes = 500
np.random.seed(2)
bits = np.random.randint(2, size = (numOFDMframes*Ni, int(np.log2(modOrder))))

symbTx = modulateGray(bits, modOrder, modType)
symbTx = pnorm(symbTx)

pilotSymb = 0.25*(max(symbTx.real) + 1j*max(symbTx.imag))
paramOFDM.pilot = pilotSymb
```

```python
sigTx = modulateOFDM(symbTx, paramOFDM)
sigTx = pnorm(sigTx)

t = np.arange(0, len(sigTx))*1/Fs
t_CP   = SpS * paramOFDM.G * (1/Fs)
t_symb = SpS * (paramOFDM.Nfft + paramOFDM.G) * (1/Fs)
```

```python
## Parametrização dos dispositivos

# Parâmetros do MZM
paramMZM = parameters()
paramMZM.Vpi = 3
paramMZM.Vb = -paramMZM.Vpi/2
paramMZM.Pin_OF = 10

# Parâmetros do canal óptico
paramChannel = parameters()
paramChannel.L = 25
paramChannel.alpha = 0.2
paramChannel.D = 16
paramChannel.Fc = 193.1e12
paramChannel.Fs = Fs

#
paramRF = parameters()
paramRF.fc_e = 7.5e9
paramRF.Pin_RF = 17
paramRF.Pin_PA = 5

# Parâmetros do fotodiodo
paramPD = parameters()
paramPD.ideal = False
paramPD.B = paramRF.fc_e + 2*Rs
paramPD.Ipd_sat = 50e-3
paramPD.Fs = Fs
```

### 2 Canal RoF

```python
# Sinal RF
sigTx_RF = np.real( sigTx * np.exp(1j*2*pi*paramRF.fc_e*t) )
G_1 = 10**( np.log10(dBm2W(paramRF.Pin_RF)/signal_power(sigTx_RF))/2)
sigTx_RF *= G_1

# Sinal óptico
Ai     = np.sqrt(dBm2W(paramMZM.Pin_OF))*np.ones(sigTx_RF.size)
sigTxo = mzm(Ai, sigTx_RF, paramMZM)

numtaps = 4096
hopt_tx = firwin(numtaps, paramRF.fc_e + 2*Rs, fs = Fs)
sigTxo = np.sqrt(signal_power(sigTxo))*pnorm(firFilter(hopt_tx, sigTxo))

sigRxo = linearFiberChannel(sigTxo, paramChannel)

# Sinal elétrico (fotocorrente)
I_Rx = photodiode(sigRxo, paramPD)
I_Rx -= I_Rx.mean()

# Sinal elétrico pós-FPF
numtaps = 4096
f1 = paramRF.fc_e - 2*Rs
f2 = paramRF.fc_e + 2*Rs
hbp_RF = firwin(numtaps, (f1, f2), pass_zero = 'bandpass', fs = Fs)

I_RF = firFilter(hbp_RF, I_Rx)

# Sina elétrico pós-PA
sigRx = hilbert(I_RF)*np.exp(-1j*2*pi*paramRF.fc_e*t)
G_2 = 10**( np.log10(dBm2W(paramRF.Pin_PA)/signal_power(sigRx))/2)
sigRx *= G_2

sigRx = powerAmplifier(sigRx)

sigRx_PA = sigRx.copy()
G_3 = 10**( np.log10(signal_power(sigTx)/signal_power(sigRx_PA))/2)
sigRx_PA *= G_3

numtaps = 4096
hlp = firwin(numtaps, Rs/1.5, fs = Fs)
#sigRx = firFilter(hlp, sigRx)

delay = finddelay(sigRx, sigTx)
sigRx = np.roll(sigRx, -delay)

rot = np.mean(sigTx/sigRx)
sigRx = rot/np.abs(rot)*sigRx
```

### 3 Preparação do conjunto de dados de treinamento

```python
SpS_nn = 4

paramDec = parameters()
paramDec.SpS_in = SpS
paramDec.SpS_out = SpS_nn

hlp = firwin(numtaps, SpS_nn*Rs/2, fs = Fs)

sigRef = decimate(pnorm(firFilter(hlp, sigTx)), paramDec).ravel()
sigIn  = decimate(pnorm(sigRx_PA), paramDec).ravel()

delay = finddelay(sigIn, sigRef)
sigIn = np.roll(sigIn, -delay)

rot = np.mean(sigRef/sigIn)
sigIn = rot/np.abs(rot)*sigIn
```

```python
sigRef = th.from_numpy(sigRef).to(device).type(th.complex64)
sigIn  = th.from_numpy(sigIn).to(device).type(th.complex64)
```

### 4 - Arquitetura de Aprendizado Indireto (ILA)

```python
def objective_ILA(trial):
    # Non-optimized param
    N = 20_000
    divByL        = 1
    trainTestFrac = 0.75
    shuffle       = False
    includeMemory = True
    augment       = True
    epochs        = 100
    batch_size    = 100
    lr            = 5e-3
    
    # Optimized param
    Nlayers = trial.suggest_int("Nlayers", 1, 3)
    Ntaps   = trial.suggest_int("Ntaps", 2, 10, step = 2)
    K       = trial.suggest_int("K", 1, 5)
    N1      = trial.suggest_int("N1", 5, 50)
    N2      = trial.suggest_int("N2", 5, 50)
    N3      = trial.suggest_int("N3", 5, 50)

    layers_full = [N1, N2, N3]
    
    layers = layers_full[0:Nlayers]
    layers.append(2)
    layers.insert(0, (K+2)*Ntaps)
    
    # Generate the model
    DPD_model = MLP(layers, activation = nn.ReLU()).to(device)
    
    loss_fn   = nn.MSELoss()
    optimizer = th.optim.Adam(DPD_model.parameters(), lr = lr)
    
    # Prepare the datasets
    train_dataloader, test_dataloader = createDatasets(sigRef[0:N], sigIn[0:N], divByL, trainTestFrac,\
                                                       batch_size, includeMemory, Ntaps, K, augment=augment)
    
    trainLoss = np.zeros(epochs)
    testLoss  = np.zeros(epochs)
    
    for t in range(epochs): 
        train_losses = train_model(train_dataloader, DPD_model, loss_fn, optimizer)
        test_losses  = test_model(test_dataloader, DPD_model, loss_fn)    
        
        trainLoss[t] = np.mean(train_losses)
        testLoss[t]  = np.mean(test_losses)


    ## Test as pre-distorter
    # Generating signal
    G_list = (G_1, G_2, G_3)
    
    paramOFDM.SpS = SpS_nn
    sigTx = modulateOFDM(symbTx, paramOFDM)
    sigTx = pnorm(sigTx)
    sigTx = th.from_numpy(sigTx).to(device).type(th.complex64)
        
    # DPD-ILA
    DPD_model.eval()
    sigTx_ILA = fitFilterNN(sigTx, DPD_model, Ntaps, K, 1, 4000, augment = augment)
    sigTx_ILA = sigTx_ILA.detach().cpu().numpy()
    P_ILA     = signal_power(sigTx_ILA)

    # Upsampling and filtering
    h_dpd = firwin(4096, 2*Rs, fs = Fs)
    sigTx_ILA = upsample(sigTx_ILA.reshape(-1,1), SpS//SpS_nn).ravel()
    sigTx_ILA = firFilter(h_dpd, sigTx_ILA)
    sigTx_ILA = np.sqrt(P_ILA)*pnorm(sigTx_ILA)

    # Channel
    sigRx_PA_DPD_ILA, symbRx_ILA = RoF_channel(sigTx_ILA, G_list, Rs, SpS, Fs, paramMZM, paramChannel, paramRF, paramPD)

    ## Performance metrics
    # NFLOPs
    Naux = 0
    for i in range(Nlayers + 1):
        Naux += 2*layers[i] + layers[i+1]*(2*layers[i] - 1)
    NFLOPs = 2 + Ntaps*(5 - K) + Naux

    # BER, SNR, EVM, MSE, ACLR
    discard = Ni
    index = np.arange(0, symbTx.size - discard)
    
    BER, _, SNR = fastBERcalc(symbRx_ILA[index], symbTx[index], modOrder, modType)
    MSE = testLoss[epochs-1]
    EVM = np.sqrt(calcEVM(symbRx_ILA[index], modOrder, modType)[0])*100
    
    freq, P_sigRx_PA_DPD_ILA = welch(pnorm(sigRx_PA_DPD_ILA)[0::SpS//SpS_nn], fs = SpS_nn*Rs, nfft = 16*1024, return_onesided = False)
    ACLR = calcACLR(P_sigRx_PA_DPD_ILA, freq, 0.5e9)
    
    # Save NN model
    th.save(DPD_model.state_dict(), file_path + f'\\NN_study_{trial.number+1}.pth')
    
    # Save parameters and performance metrics
    metrics_array = np.array([ACLR, MSE, BER[0], EVM, SNR[0], NFLOPs, Nlayers, N1, N2, N3, Ntaps, K])
    np.savetxt(file_path + f'\\metrics_{trial.number+1}.txt', metrics_array, fmt = '%f')
    
    return EVM, ACLR, NFLOPs
```

```python
n_trials  = 1000
n_studies = 1
save = True

for i in range(n_studies):
    print(f"Study {i+1}:")
    study_ILA = optuna.create_study(directions=["minimize", "minimize", "minimize"], sampler = optuna.samplers.TPESampler())
    study_ILA.optimize(objective_ILA, n_trials = n_trials, timeout = 60*60*15, show_progress_bar = True)
    
    complete_trials   = study_ILA.get_trials(deepcopy = False, states = [TrialState.COMPLETE])
    n_complete_trials = len(complete_trials)
    
    study_N1      = np.zeros(n_complete_trials)
    study_N2      = np.zeros(n_complete_trials)
    study_N3      = np.zeros(n_complete_trials)
    study_Ntaps   = np.zeros(n_complete_trials)
    study_K       = np.zeros(n_complete_trials)
    study_Nlayers = np.zeros(n_complete_trials)
    
    study_EVM    = np.zeros(n_complete_trials)
    study_ACLR   = np.zeros(n_complete_trials)
    study_NFLOPs = np.zeros(n_complete_trials)
    
    for j in range(n_complete_trials):
        trials_params    = complete_trials[j].params
        study_EVM[j]     = complete_trials[j].values[0]
        study_ACLR[j]     = complete_trials[j].values[1]
        study_NFLOPs[j]  = complete_trials[j].values[2]
        
        study_N1[j]      = trials_params["N1"]
        study_N2[j]      = trials_params["N2"]
        study_N3[j]      = trials_params["N3"]
        study_Ntaps[j]   = trials_params["Ntaps"]
        study_K[j]       = trials_params["K"]
        study_Nlayers[j] = trials_params["Nlayers"]

    if save:
        np.savetxt(file_path + f'\\ILA_study{i+1}_N1.txt', study_N1, fmt = '%f')
        np.savetxt(file_path + f'\\ILA_study{i+1}_N2.txt', study_N2, fmt = '%f')
        np.savetxt(file_path + f'\\ILA_study{i+1}_N3.txt', study_N3, fmt = '%f')
        np.savetxt(file_path + f'\\ILA_study{i+1}_Ntaps.txt', study_Ntaps, fmt = '%f')
        np.savetxt(file_path + f'\\ILA_study{i+1}_K.txt', study_K, fmt = '%f')
        np.savetxt(file_path + f'\\ILA_study{i+1}_Nlayers.txt', study_Nlayers, fmt = '%f')
        np.savetxt(file_path + f'\\ILA_study{i+1}_ACLR.txt', study_ACLR, fmt = '%f')
        np.savetxt(file_path + f'\\ILA_study{i+1}_EVM.txt', study_EVM, fmt = '%f')
        np.savetxt(file_path + f'\\ILA_study{i+1}_NFLOps.txt', study_NFLOPs, fmt = '%f') 
```

```python
def get_pareto(f1, f2, n_trials):
    
    solutions = np.hstack( (f1.reshape((n_trials, 1)), f2.reshape((n_trials, 1))) )
    pareto_solutions = []
    
    for i, s in enumerate(solutions):
        test = np.where( (s[0] > solutions[:,0]) * (s[1] > solutions[:,1]) )[0]
        
        if not(len(test)):
            pareto_solutions.append(tuple(s))

    pareto_solutions.sort()
    return np.array(pareto_solutions)
```

```python
def get_pareto(f1, f2, f3, n_trials):
    
    solutions = np.hstack( (f1.reshape((n_trials, 1)), f2.reshape((n_trials, 1)), f3.reshape((n_trials, 1))) )
    pareto_solutions = []
    
    for i, s in enumerate(solutions):
        test = np.where( (s[0] > solutions[:,0]) * (s[1] > solutions[:,1]) * (s[2] > solutions[:,2])  )[0]
        
        if not(len(test)):
            pareto_solutions.append(tuple(s))

    pareto_solutions.sort()
    return np.array(pareto_solutions)
```

```python
pareto_solutions = get_pareto(EVM, ACLR, NFLOPs, n_trials)
```
